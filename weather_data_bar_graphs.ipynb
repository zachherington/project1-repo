```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
file_to_load = "ny_weather_0119-0619.csv"
weather_df = pd.read_csv(file_to_load)
weather_df
```

```python
precipitation_df = weather_df.sort_values(by="Precipitation",ascending=False)
precipitation_df
```

```python
final_precipitation_df = precipitation_df.iloc[:10]
```

```python
final_precipitation_df.plot.bar(x="date",y="Precipitation")
plt.ylabel("Precipitation")
plt.title("Top Ten Rain Days")
plt.xticks(rotation=30, horizontalalignment="center")
```

```python
snow_df = weather_df.sort_values(by="Snowfall",ascending=False)
```

```python
final_snow_df = snow_df.iloc[:10]
```

```python
final_snow_df.plot.bar(x="date",y="Snowfall")
plt.ylabel("Snowfall")
plt.title("Top Ten Snow Days")
plt.xticks(rotation=30, horizontalalignment="center")
```

```python
wind_df = weather_df.sort_values(by="Wind Speed",ascending=False)
```

```python
final_wind_df = wind_df.iloc[:10]
```

```python
final_wind_df.plot.bar(x="date",y="Wind Speed")
plt.ylabel("Wind Speed")
plt.title("Top Ten Wind Days")
plt.xticks(rotation=30, horizontalalignment="center")
```
